```python
# Find missing values
missing_values = df.isnull().sum()

# Display the results
print(missing_values)
```

in abouve output we see some columns are contains null values. the missing values in ADDRESSLINE2, STATE, POSTALCODE and TERRITORY.

```python
# 1. Fix TERRITORY values
df['TERRITORY'] = df['TERRITORY'].replace('NaN', 'North America')
# this is not working because NaN is not any value that why we can't replace it
```

```python
df['TERRITORY'] = df['TERRITORY'].fillna('North America')
```

```python
print(df['TERRITORY'])
```

```python
# 2. Drop 'ADDRESSLINE2' (with correct casing)
df.drop(columns=['ADDRESSLINE2'], inplace=True)
```

```python
# First, sort the DataFrame by the grouping columns
df = df.sort_values(['CITY', 'COUNTRY'])

# Fill missing 'state' values by propagating non-null values within each city and country group
df['STATE'] = df.groupby(['CITY', 'COUNTRY'])['STATE'].transform(lambda s: s.ffill().bfill())

# Similarly, fill missing 'postalcode' values
df['POSTALCODE'] = df.groupby(['CITY', 'COUNTRY'])['POSTALCODE'].transform(lambda s: s.ffill().bfill())
```

```python
print(df[df['STATE'].isnull()][['CITY', 'COUNTRY', 'STATE']])
print(df[df['POSTALCODE'].isnull()][['CITY', 'COUNTRY', 'POSTALCODE']])
```

```python
# Find duplicates
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# change the ORDERDATE column format as dd/mm/yyyy
# Step 1: Convert the ORDERDATE column to datetime format (if it's not already)
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')

# Step 2: Format it to 'dd/mm/yyyy' string format
df['ORDERDATE'] = df['ORDERDATE'].dt.strftime('%d/%m/%Y')
```

```python
# 1. Normalize PHONE numbers to international format (e.g., +1-123-456-7890)
import re

def format_phone(phone):
    digits = re.sub(r'\D', '', str(phone))  # Keep only digits
    if len(digits) == 10:
        return f'+1-{digits[0:3]}-{digits[3:6]}-{digits[6:]}'
    elif len(digits) == 11 and digits.startswith('1'):
        return f'+1-{digits[1:4]}-{digits[4:7]}-{digits[7:]}'
    else:
        return phone  # leave as-is if format unknown

df['PHONE'] = df['PHONE'].apply(format_phone)

# 2. STATE column: Uppercase
df['STATE'] = df['STATE'].str.upper().str.strip()

# 3. POSTALCODE: Convert to string (preserve leading 0s)
df['POSTALCODE'] = df['POSTALCODE'].astype(str).str.zfill(5)

# 4. Title Case all object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.title().str.strip()
```

```python
df.rename(columns={
    'ORDERNUMBER': 'ORDER_ID',
    'QUANTITYORDERED': 'QUANTITY',
    'PRICEEACH': 'PRICE_EACH',
    'ORDERLINENUMBER': 'LINE_NUMBER',
    'SALES': 'TOTAL_SALES',
    'ORDERDATE': 'ORDER_DATE',
    'STATUS': 'ORDER_STATUS',
    'PRODUCTLINE': 'PRODUCT_LINE',
    'CUSTOMERNAME': 'CUSTOMER_NAME',
    'CONTACTLASTNAME': 'LAST_NAME',
    'CONTACTFIRSTNAME': 'FIRST_NAME',
    'DEALSIZE': 'DEAL_SIZE',
    'PRODUCTCODE': 'PRODUCT_CODE',
    'MSRP': 'MSRP'
}, inplace=True)
```

```python
# Fix dates
df['ORDER_DATE'] = pd.to_datetime(df['ORDER_DATE'], errors='coerce', dayfirst=True)

# Postal codes as string (to keep leading 0s)
df['POSTALCODE'] = df['POSTALCODE'].astype(str).str.zfill(5)

# Convert categories
df['ORDER_STATUS'] = df['ORDER_STATUS'].astype('category')
df['PRODUCT_LINE'] = df['PRODUCT_LINE'].astype('category')
```

it looks like this csv is a combination of multiple csv's like order_details, customer_details, product_details etc., so we will split the dataset depends on catagry.

```python
# 1. Orders CSV
orders_df = df[['ORDER_ID', 'ORDER_DATE', 'ORDER_STATUS', 'TOTAL_SALES', 'DEAL_SIZE', 'TERRITORY']]
orders_df.to_csv('orders.csv', index=False)

# 2. Order Details CSV
order_details_df = df[['ORDER_ID', 'QUANTITY', 'PRICE_EACH', 'LINE_NUMBER', 'PRODUCT_CODE', 'MSRP']]
order_details_df.to_csv('order_details.csv', index=False)

# 3. Products CSV (deduplicated by productcode)
products_df = df[['PRODUCT_CODE', 'PRODUCT_LINE', 'MSRP']].drop_duplicates()
products_df.to_csv('products.csv', index=False)

# 4. Customers CSV
customers_df = df[['CUSTOMER_NAME', 'PHONE', 'ADDRESSLINE1', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY']]
customers_df.to_csv('customers.csv', index=False)

# 5. Contacts CSV
contacts_df = df[['CUSTOMER_NAME', 'FIRST_NAME', 'LAST_NAME']]
contacts_df.to_csv('contacts.csv', index=False)
```
